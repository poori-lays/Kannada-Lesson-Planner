```python
import torch
from datasets import load_dataset
from transformers import (
    T5Tokenizer,
    LongT5ForConditionalGeneration,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)

# -----------------------------
# CONFIG
# -----------------------------
MODEL_NAME = "google/mt5-small"
DATA_PATH = "train.json"
OUTPUT_DIR = "longt5_kannada_model"

MAX_INPUT_LENGTH = 1024#4096
MAX_OUTPUT_LENGTH = 512#1536

BATCH_SIZE = 1
EPOCHS = 20
LEARNING_RATE = 3e-4

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# -----------------------------
# LOAD DATA
# -----------------------------
dataset = load_dataset("json", data_files=DATA_PATH, split="train")

# Use same data for train & eval (sanity check)
train_data = dataset
eval_data = dataset

# -----------------------------
# TOKENIZER & MODEL
# -----------------------------
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)

model = LongT5ForConditionalGeneration.from_pretrained(
    MODEL_NAME
).to(device)
model.gradient_checkpointing_enable()
# -----------------------------
# PREPROCESS
# -----------------------------
def preprocess(examples):
    model_inputs = tokenizer(
        examples["input"],
        truncation=True,
        max_length=MAX_INPUT_LENGTH
    )

    labels = tokenizer(
        examples["output"],
        truncation=True,
        max_length=MAX_OUTPUT_LENGTH
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
'''
def preprocess(examples):
    inputs = tokenizer(
        examples["input"],
        truncation=True,
        padding="max_length",
        max_length=MAX_INPUT_LENGTH
    )
    outputs = tokenizer(
        examples["output"],
        truncation=True,
        padding="longest",
        #max_length=MAX_OUTPUT_LENGTH
    )

    labels = []
    for label in outputs["input_ids"]:
        labels.append(
            [(tok if tok != tokenizer.pad_token_id else -100) for tok in label]
        )

    inputs["labels"] = labels
    return inputs
'''
train_data = train_data.map(preprocess, batched=True, remove_columns=train_data.column_names)
eval_data = eval_data.map(preprocess, batched=True, remove_columns=eval_data.column_names)

# -----------------------------
# TRAINING ARGS
# -----------------------------
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    learning_rate=LEARNING_RATE,
    num_train_epochs=EPOCHS,
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_steps=1,
    fp16=True,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=eval_data,
    data_collator=DataCollatorForSeq2Seq(tokenizer, model)
)

# -----------------------------
# TRAIN
# -----------------------------
trainer.train()

print("✅ Training complete")
```

```python
import torch
import json
from transformers import T5Tokenizer, LongT5ForConditionalGeneration

# -----------------------------
# CONFIG
# -----------------------------
MODEL_DIR = "longt5_kannada_model"   # same OUTPUT_DIR you used in training
MAX_INPUT_LENGTH = 4096
MAX_OUTPUT_LENGTH = 1536

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# -----------------------------
# LOAD MODEL & TOKENIZER
# -----------------------------
tokenizer = T5Tokenizer.from_pretrained(MODEL_DIR)
model = LongT5ForConditionalGeneration.from_pretrained(MODEL_DIR).to(device)
model.eval()

# -----------------------------
# LOAD INPUT (from train.json)
# -----------------------------
with open("train.json", "r", encoding="utf-8") as f:
    data = json.load(f)

input_text = data[0]["input"]

# -----------------------------
# TOKENIZE
# -----------------------------
inputs = tokenizer(
    input_text,
    return_tensors="pt",
    truncation=True,
    max_length=MAX_INPUT_LENGTH
).to(device)

# -----------------------------
# GENERATE
# -----------------------------
with torch.no_grad():
    output_ids = model.generate(
        **inputs,
        max_length=MAX_OUTPUT_LENGTH,
        min_length=200,
        num_beams=1,        # can set to 1 if GPU memory is tight
        do_sample=False
    )

# -----------------------------
# DECODE
# -----------------------------
generated_text = tokenizer.decode(
    output_ids[0],
    skip_special_tokens=True
)

print("\n===== GENERATED LESSON PLAN =====\n")
print("Generated token IDs:", output_ids[0][:50])
print(tokenizer.decode(output_ids[0], skip_special_tokens=True))
print(generated_text)
```
